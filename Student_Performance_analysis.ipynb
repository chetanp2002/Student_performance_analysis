```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_score.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Drop unnamed column
df = df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
# gender distribution
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# from the above chart we have analysed that:
#     the number of females in the dat ais more tjan the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot=True)
plt.title("Relationship between Parent's Education and Student's scoreRelationship between Parent's Education and Student's score")
plt.show()
```

```python
# from the above chart we have analysed that:
#     the education of the parents have a good impact on their scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot=True)
plt.title("Relationship between Parent's Marital status and Student's score")
plt.show()
```

```python
# from the above chart we have analysed that:
#     there is no/neglible impact on the student score due to their marital status
```
